### Univariate Analysis

#### Distplot

```python
sns.distplot(aids_df["karnof"],kde=True,bins=10)
```

The distplot shows patient Karnofsky scores. Most scores are around 85 to 100, indicating good health

#### Countplot

```python
sns.countplot(x="infected",data=aids_df)
```

This countplot how many of them are infected and non‑infected patients. About 1,600 are not infected, while around 500 are infected

### Bivariate Analysis

#### Boxplot

```python
sns.boxplot(x="infected",y="cd40",data=aids_df)
```

This box plot shows CD4 levels for infected and non‑infected patients. Infected patients usually have lower counts, and both groups include some high outliers.

#### Jointplot

```python
sns.jointplot(x="cd40",y="cd420",data=aids_df,kind="reg")
```

The jointplot shows CD4 counts at week 0 and week 20. People who start with high CD4 counts usually stay high

####  Scatterplot

```python
sns.scatterplot(x="age", y="cd40", data=aids_df)
plt.xlabel("Age")
plt.ylabel("CD4 Count (cd40)")
plt.show()
```

Most patients are 20–50 years old with CD4 counts between 200–600

#### Violinplot

```python
sns.violinplot(x="gender",y="cd40",data=aids_df)
```

This plot compares CD4 counts (cd40) between two gender groups but female(0)  has a higher median.

### Multivariate Analysis

#### Heatmap

```python
drop_cols=['time','trt','age','wtkg','hemo','homo','drugs','race','gender','offtrt','symptom','oprior','treat','karnof']
aids_data=aids_df.drop(columns=drop_cols)
```

```python
plt.figure(figsize=(10,10))
aids_mx=(aids_data.corr())
sns.heatmap(aids_mx,annot=True,cmap='Blues')
```

#### Pieplot

```python
trt_labels = {0: "ZDV only", 1: "ZDV + ddl", 2: "ZDV + zal", 3: "ddl only"}
aids_df['trt_label'] = aids_df['trt'].map(trt_labels)
trt_counts = aids_df['trt_label'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(trt_counts, labels=trt_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Treatment Group Distribution")
plt.show()
```

#### Pairplot

```python
sns.pairplot(
    aids_data,
    hue="infected",
    palette="Set1",
    diag_kind="kde", 
    kind="reg",
    plot_kws={"scatter_kws": {"s": 70}}
)
plt.show()
```

## Outlier Handling

```python
plt.figure(figsize=(20,10))
aids_df.boxplot()
```

```python
for i in aids_df.select_dtypes(include='number').columns:
    q1 = aids_df[i].quantile(0.25)
    q3 = aids_df[i].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    aids_df[i] = aids_df[i].clip(lower, upper)
```

```python
plt.figure(figsize=(20,10))
aids_df.boxplot()
```
