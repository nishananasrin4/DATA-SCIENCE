```python
df["Location"] = df["Location"].fillna("unknown")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
import string
punctuation=string.punctuation
nltk.download("stopwords")
from nltk.corpus import stopwords
sp=set(stopwords.words("english"))
nltk.download("punkt")
lemmatizer = WordNetLemmatizer()

def clean(text):
    words=word_tokenize(text)
    words=[word for word in words if word not in sp and word not in punctuation]
    words=[lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
```

```python
import re
```

```python
def preprocess_text(text): 
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    text=re.sub('\w*\d\w*','',text)
    return text
```

```python
df['OriginalTweet']=df['OriginalTweet'].apply(preprocess_text)
df['OriginalTweet']=df['OriginalTweet'].apply(clean)
```

```python
df
```

```python
df['Sentiment'] = df['Sentiment'].replace({
    'Extremely Negative': -1,
    'Negative': -1,
    'Neutral': 0,
    'Positive': 1,
    'Extremely Positive': 1
})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
en_cols=['UserName', 'ScreenName', 'Location', 'TweetAt']
```

```python
for i in en_cols:
    df[i]=le.fit_transform(df[i])
```

```python
x=df['OriginalTweet']
```

```python
x
```

```python
y=df['Sentiment']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vc=CountVectorizer()
```

```python
xtrain_vc=vc.fit_transform(x_train)
xtest_vc=vc.transform(x_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr=LogisticRegression()
```

```python
model_lr.fit(xtrain_vc,y_train)
```

```python
ypred_lr=model_lr.predict(xtest_vc)
```

```python
ypred_lr
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_lr=accuracy_score(y_test,ypred_lr)
```

```python
acc_lr
```
