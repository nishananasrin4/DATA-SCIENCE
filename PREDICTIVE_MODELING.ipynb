## OUTLIER HANDLING

```python
plt.figure(figsize=(20,10))
df.boxplot()
```

```python
for i in df.select_dtypes(include='number').columns:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    df[i] = df[i].clip(lower, upper)
plt.figure(figsize=(20,10))
df.boxplot()
```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
metric_k=[]
```

```python
neighbors=np.arange(2,14)
```

```python
for i in neighbors:
    model_knn=KNeighborsRegressor(n_neighbors=i,metric='euclidean')
    model_knn.fit(x_train,y_train)
    knn_p=model_knn.predict(x_test)
    r2=r2_score(y_test,knn_p)
    metric_k.append(r2)
```

```python
plt.plot(neighbors,metric_k,'*-')
plt.title('Graph to Find K')
plt.xlabel('k')
plt.ylabel('metric_k')
plt.show()
```

```python
model_knn=KNeighborsRegressor(n_neighbors=14,metric='euclidean')
model_knn.fit(x_train,y_train)
y_pred_knn=model_knn.predict(x_test)
r2_knn=r2_score(y_test,y_pred_knn)
```

## HYPERPARAMETER TUNING

#### RandomizedSearchCV

```python
import lightgbm as lgb
model=lgb.LGBMRegressor(verbose=0)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [200, 500],
    'learning_rate': [0.01, 0.05, 0.1],
    'num_leaves': [31, 50, 100],
    'max_depth': [-1, 5, 10],
    'min_child_samples': [20, 50],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
random_search = RandomizedSearchCV(model, param_distributions=param_grid,cv=5,verbose=0) 
```

```python
random_search.fit(x_train, y_train) 
```

```python
print("Best Parameters:",random_search.best_params_)
```

```python
print("Best Score (Accuracy):", random_search.best_score_) 
```
